# Explore: image search for products

```python
import sys
sys.path.append("../src")

from read_excel import load_products
from query_builder import build_query, build_fallback_query
from search_ddgs import search_images

from IPython.display import HTML, display
```

## 1. Load the first 10 products

```python
products = load_products()
products
```

## 2. Pick a product and build its query

```python
ROW = 0
row = products.iloc[ROW]
query = build_query(row)
print(f"Product : {row['Nombre']} ({row['Marca']})")
print(f"Query   : {query}")
print(f"Fallback: {build_fallback_query(row)}")
```

## 3. Search DuckDuckGo images

```python
candidates = search_images(query)
print(f"{len(candidates)} candidates")
candidates[0]
```

## 4. See the candidates

```python
def show_candidates(cands):
    cards = []
    for i, c in enumerate(cands):
        cards.append(f"""
        <div style='display:inline-block;width:170px;margin:6px;vertical-align:top;
                    border:1px solid #ccc;border-radius:6px;padding:6px;text-align:center'>
          <a href='{c['page_url']}' target='_blank'>
            <img src='{c['thumbnail_url']}' style='max-width:150px;max-height:150px'>
          </a>
          <div style='font-size:11px;margin-top:4px'>
            <b>#{i}</b> {c['width']}x{c['height']}<br>
            {c['source_site'][:25]}<br>
            <span style='color:#777'>{c['title'][:40]}</span>
          </div>
        </div>""")
    display(HTML("<div>" + "".join(cards) + "</div>"))

show_candidates(candidates)
```

## 5. Testing the query


```python
test_query = build_fallback_query(products.iloc[8])
print(test_query)
show_candidates(search_images(test_query))
```
